```python
import pandas as pd
import numpy as np
```

```python
chn = []
sub_chn = []
office = []
prod = []
price = []
cartons = []
weight = []
sales = []
period = []

channel_def = {
    "K": ["KA", "KB", "KC"],
    "E": ["EE"],
    "G": ["GB"],
    "M": ["MA", "MB", "MC"],
    "P": ["PA"],
    "A": ["AA", "AB", "AC", "AD", "AE", "AG", "AH"],
    "S": ["SA", "SB", "SC"],
    "D": ["D0", "D1"]
}

group_def = {
    "K": ["CE1", "CE8", "CEB", "CEI", "CEK"],
    "E": ["CEE"],
    "G": ["CE2", "CE7", "CEB", "CEI", "CEL"],
    "M": ["CE2", "CE7", "CEA", "CEI", "CEL"],
    "P": ["CEE"],
    "A": ["CE3", "CE8", "CEA", "CEI", "CEK"],
    "S": ["CEE", "CE8", "CEB", "CEI", "CEK"],
    "D": ["CE8"]
}

workingdays = {
    "2023/01": 25,
    "2023/02": 23,
    "2023/03": 26,
    "2023/04": 26,
    "2023/05": 25,
    "2023/06": 25,
    "2023/07": 24,
    "2023/08": 23,
    "2023/09": 24,
    "2023/10": 25,
    "2023/11": 25,
    "2023/12": 24
}

total_days = sum(workingdays.values())

houses = {
    "Taipei": 0.371658314156427,
    "Hshinchu": 0.173392762825309,
    "Taichung": 0.171451484425119,
    "Tainan": 0.115679015622988,
    "Kaohsiung": 0.167818422970157
}

people = {
    "Taipei": 0.351130615776822,
    "Hshinchu": 0.179059724667336,
    "Taichung": 0.188393578337422,
    "Tainan": 0.119003623302728,
    "Kaohsiung": 0.162412457915691
}

weighted ={
    "Taipei": houses["Taipei"]*0.5+people["Taipei"]*0.5,
    "Hshinchu": houses["Hshinchu"]*0.5+people["Hshinchu"]*0.5,
    "Taichung": houses["Taichung"]*0.5+people["Taichung"]*0.5,
    "Tainan": houses["Tainan"]*0.5+people["Tainan"]*0.5,
    "Kaohsiung": houses["Kaohsiung"]*0.5+people["Kaohsiung"]*0.5
}

max_value = max(weighted.values())
max_key = max(weighted, key=weighted.get)

if sum(weighted.values()) > 1:
    weighted[max_key] = max_value-(sum(weighted.values())-1)
elif sum(weighted.values()) == 1:
    pass
else:
    weighted[max_key] = max_value+(1-sum(weighted.values()))

share_offices = {
    "CEE": weighted["Taipei"],
    "CE1": weighted["Taipei"],
    "CE2": weighted["Taipei"],
    "CE3": weighted["Taipei"],
    "CE7": weighted["Hshinchu"],
    "CE8": weighted["Hshinchu"],
    "CEA": weighted["Taichung"],
    "CEB": weighted["Taichung"],
    "CEI": weighted["Tainan"],
    "CEK": weighted["Kaohsiung"],
    "CEL": weighted["Kaohsiung"]
}

employee_units ={
    "HD": 347,
    "Audit": 17,
    "Paper_Container": 157,
    "Banqiao": 188, 
    "House_Product": 117,
    "Dayuan_paper_container": 341,
    "Miaoli": 155,
    "Zhubei": 628,
    "Dayuan": 341,
    "Hshinchu": 114,
    "Taichung": 134,
    "Industrial_Paper": 89,
    "Houli": 799,
    "Dalin": 189,
    "Yanchao": 158
}

S_share_offices = {
    "CEE": (employee_units["HD"]+employee_units["Audit"]+employee_units["Paper_Container"]+employee_units["Banqiao"]+employee_units["House_Product"])/sum(employee_units.values()),
    "CE8": (employee_units["Dayuan_paper_container"]+employee_units["Miaoli"]+employee_units["Zhubei"]+employee_units["Dayuan"]+employee_units["Hshinchu"])/sum(employee_units.values()),
    "CEB": (employee_units["Taichung"]+employee_units["Industrial_Paper"]+employee_units["Houli"])/sum(employee_units.values()),
    "CEI": (employee_units["Dalin"])/sum(employee_units.values()),
    "CEK": (employee_units["Yanchao"])/sum(employee_units.values())
}

car_to_wei = {
    "1091100": 0.007920,
    "1091110": 0.010240,
    "1091130": 0.009201, 
    "1091140": 0.010641,
    "1091150": 0.009227,
    "1091160": 0.010643,
    "1091190": 0.010237,
    "1091200": 0.008960,
    "1091300": 0.006257,
    "1091320": 0.006981,
    "1091500": 0.007011,
    "1091501": 0.011646,
    "1091510": 0.013845,
    "1091520": 0.007787,
    "1091521": 0.009984, 
    "1091530": 0.014553,
    "1091700": 0.003034,
    "10W1110": 0.007125, 
    "10W1111": 0.005405,
    "10W1120": 0.009358,
    "10W1140": 0.000000,
    "10W1160": 0.011842,
    "10W1170": 0.009103,
    "10W1171": 0.007986,
    "10W1190": 0.009061,
    "10W1200": 0.006359,
    "10W1210": 0.007248,
    "10W1220": 0.005378,
    "10W1520": 0.006829,
    "10W3110": 0.006190,
    "10W3120": 0.005510,
    "10W3220": 0.006710,
    "10W1Z00": 0.013000,
    "10W2230": 0.006970,
    "10W2410": 0.005730,
    "10W5110": 0.007070,
    "10W5120": 0.008400,
    "10W5140": 0.004800,
    "10W8100": 0.009390,
    "10W8110": 0.010800   
}
```

```python
sheet = pd.ExcelFile("Goals_2023.xlsx")
group_A = ["K", "M", "A", "S", "D"]
group_B = ["E", "G", "P"]

for s_name in sheet.sheet_names:
    if s_name in group_A:
        excel_data_df = pd.read_excel("Goals_2023.xlsx", sheet_name=s_name, usecols=("A, C, E:Z"))
    else:
        excel_data_df = pd.read_excel("Goals_2023.xlsx", sheet_name=s_name, usecols=("A, C:D"))
    nonnull_data = excel_data_df.dropna()
    data_list = nonnull_data.values.tolist()
    
    for data in data_list:
        key_prod = str(data[0])
        prod += len(channel_def[s_name])*len(group_def[s_name])*len(workingdays)*[data[0]]
        price += len(channel_def[s_name])*len(group_def[s_name])*len(workingdays)*[data[1]]
        chn += len(channel_def[s_name])*len(group_def[s_name])*len(workingdays)*["2"+s_name]

        for subchn in channel_def[s_name]:
            sub_chn += len(group_def[s_name])*len(workingdays)*[subchn]

        for car in data[2:]:
            for workingday in workingdays:
                period += len(group_def[s_name])*[workingday]

                for group in group_def[s_name]:
                    office.append(group)
                    
                    if s_name not in ["S", "D"]:
                        if len(group_def[s_name]) > 1:
                            cartons.append(workingdays[workingday]/total_days*share_offices[group]*car)
                            sales.append(workingdays[workingday]/total_days*share_offices[group]*car*data[1])
                            weight.append(workingdays[workingday]/total_days*share_offices[group]*car*car_to_wei[key_prod])
                        else:
                            cartons.append(workingdays[workingday]/total_days*car)
                            sales.append(workingdays[workingday]/total_days*car*data[1])
                            weight.append(workingdays[workingday]/total_days*car*car_to_wei[key_prod])
                    elif s_name in ["S"]:
                        cartons.append(workingdays[workingday]/total_days*S_share_offices[group]*car)
                        sales.append(workingdays[workingday]/total_days*S_share_offices[group]*car*data[1])
                        weight.append(workingdays[workingday]/total_days*S_share_offices[group]*car*car_to_wei[key_prod])
                    else:
                        weight.append(round(workingdays[workingday]/total_days*car,2))
                        sales.append(round(workingdays[workingday]/total_days*car*data[1],2))
                        cartons.append(workingdays[workingday]/total_days*car*0)
```

```python
data = {
    "Product": prod,
    "Price": price,
    "Channel": chn,
    "SubChannel": sub_chn,
    "Year/Month": period,
    "Office": office,   
    "Cartons": cartons,
    "Sales": sales,
    "Weight": weight
}

df_converted = pd.DataFrame(data)
```
